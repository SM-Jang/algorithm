```python
import sys

# input = sys.stdin.readline
INF = int(1e9)

# 노드, 간선
n,m = map(int, input().split())
start = int(input())

# 각 노드에 연결되어 있는 노드에 대한 정보를 담는 리스트
graph = [ [] for i in range(n+1) ]
# 방문한 적이 있는지 체크
visited = [False]*(n+1)
# start노드에서 각 노드에 대한 최적의 거리
distance = [INF]*(n+1)

# 간선 정보
for _ in range(m):
    a, b, c = list(map(int, input().split(' ')))
    # a 번 노드에서 b번 노드로 가는데 비용이 c
    graph[a].append((b,c))
    
def get_smallest_node():
    """
    방문하지 않은 노드 중에서, 가장 최단 거리가 짧은 노드의 번호를 반환
    """
    min_value=INF
    index=0 # 가장 최단 거리가 짧은 노드(인덱스)
    for i in range(1, n+1):
        if distance[i] < min_value and not visited[i]:
            min_value=distance[i]
            index = i
            
    return index

def dijkstra(start):
    # 시작 노드에 대해서 초기화
    distance[start] = 0
    visited[start] = True
    
    for j in graph[start]:
        distance[j[0]] = j[1]
        # j = (b,c): node start -> node b까지 최단 거리를 c로
    
    for i in range(n-1):
        # 현재 최단 거리가 가장 짧은 노드를 꺼내서, 방문 처리
        now = get_smallest_node()
        visited[now] = True

        # 현재 노드와 연결된 다른 노드를 확인
        for j in graph[now]:
            cost = distance[now]+j[1]
            if cost<distance[j[0]]:
                distance[j[0]] = cost
                
dijkstra(start)

for i in range(1, n+1):
    if distance[i] == INF:
        print("Infinity")
    else:
        print(distance[i])
```

```python
# 개선된 다익스트라 알고리즘
import heapq

INF = int(1e9)

# 노드, 간선
n, m = map(int, input().split())
# 시작 노드
start = int(input())

# 각 노드끼리 연결되어 있는 노드 정보
graph = [ [] for _ in range(n+1) ]

# 최적의 거리를 기록 d[node]: start->node optimal distance 
distance = [INF]*(n+1)

# 간선 정보 입력 (dist, node)
for _ in range(m):
    node1, dist, node2 = map(int, input().split())
    graph[node1].append((dist, node2))
    
def dijkstra(start):
    q = [] # 힙 큐를 만드는 리스트
    
    # 시작 노드로 가기 위한 최단 경로는 0으로 설정하여 큐에 삽입
    heapq.heappush(q, (0, start)) # dist, node 정보 푸시
    distance[start] = 0
    
    while q: # 
        dist, now = heapq.heappop(q)

        if distance[now] < dist: continue
        for i in graph[now]:
            cost = dist + i[1]
            
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))
                
                
dijkstra(start)
print(distance)
```

```python
import heapq
import sys

INF = int(1e9)

# 노드, 간선, 시작 노드
n, m, start = map(int, input().split())
# 각 노드에 연결된 노드 정보
graph = [[] for _ in range(n+1)]
# 최단 거리 테이블
distance = [INF]*(n+1)


for i in range(1, m+1):
    # x->y: z
    x, y, z = map(int, input().split())
    graph[x].append((y,z)) # 우선 순위 큐에서 사용
    
def dijkstra(start):
    
    q = []
    heapq.heappush(q, (0, start)) # dist, node
    
    while q:
        dist, now = heapq.heappop(q)
        # 이미 지난 노드
        if distance[now] < dist: continue
        
        # 최소 거리의 인접 노드들 살펴보기
        for i in graph[now]:
            cost = dist + i[1] 
            # 거쳐간 거리와 직접 거리 비교
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))
                
                
dijkstra(start)

print(distance)
```
